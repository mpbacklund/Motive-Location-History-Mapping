# July 2022 - July 2023

Data was pulled from Geotab for stop events lasting longer than 5 minutes and less than 24 hours. This effectively eliminated stop events where a truck stopped in traffic and removed outliers for trucks that were parked long-term that skewed the data. 

* Total stops 11,162,801
* Median Stop Duration: 19 min
* Mean Stop Duration: 1 hr 50 mins
* Median distance between stops: 1.4 miles
* Mean distance between stops: 6.5 miles





There are 2 main local locations that trucks are stopping at. The first is at RSD Container Yard Services. Trucks frequently park in front of the yard for an average of 2 hours. In addition, trucks often park in the yard for an average of 30-70 minutes.

The second main location trucks stop at is a nearby yard at 5220 West 700 South. Trucks are stopping here for an average of 1-4 hours

The data also shows a good amount of trucks going to Salt Lake Diesel Service at 3763 West 700 South, where they spend about an hour on average. While many trucks go to this location, it is probably not a good candidate for electric vehicle chargers because the trucks are actively worked on there.
